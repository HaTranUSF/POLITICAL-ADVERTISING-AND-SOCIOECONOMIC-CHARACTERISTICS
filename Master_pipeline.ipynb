```python
# 1. Load the datasets
inc = pd.read_csv('income_panel_2020_2024.csv')
pov = pd.read_csv('poverty_panel_2020_2024.csv')
edu = pd.read_csv('education_panel_2020_2024.csv')
pop = pd.read_csv('population_panel_2020_2024.csv')
abbr = pd.read_csv('data-map-state-abbreviations.csv')

# 2. Clean whitespace
for df in [inc, pov, edu, pop]:
    df['State'] = df['State'].str.strip()
abbr['Name'] = abbr['Name'].str.strip()
abbr['Abbreviation'] = abbr['Abbreviation'].str.strip()

# 3. Merge Census Indicators on State and Year
master_df = inc.merge(pov, on=['State', 'Year'], how='outer')
master_df = master_df.merge(edu, on=['State', 'Year'], how='outer')
master_df = master_df.merge(pop, on=['State', 'Year'], how='outer')

# 4. Add State Abbreviations and swap names
# We link the full name (State) to the abbreviations file (Name)
master_df = master_df.merge(abbr, left_on='State', right_on='Name', how='left')

# Replace the full name 'State' column with the 2-letter 'Abbreviation'
master_df['State'] = master_df['Abbreviation']

# 5. Clean up redundant columns
# Drop the extra columns from the merge and the original abbreviation column
master_df = master_df.drop(columns=['Name', 'Abbreviation'])

# 6. Reorder: Move 'State' to the first position
cols = ['State'] + [c for c in master_df.columns if c != 'State']
master_df = master_df[cols]

# 7. Sort and Save
master_df = master_df.sort_values(['State', 'Year'])
master_df.to_csv('master_socioeconomic_panel.csv', index=False)
```
